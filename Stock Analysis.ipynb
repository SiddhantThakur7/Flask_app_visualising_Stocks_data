```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure, show, output_file
from bokeh.embed import components
from bokeh.resources import CDN


start = datetime.datetime(2020, 6, 30)
end = datetime.datetime.now()

df = data.DataReader(name = "GOOG", data_source="yahoo", start = start, end = end)
df

df1 = df[df.Close > df.Open]
df2 = df[df.Close < df.Open]

p =figure(x_axis_type = 'datetime', width = 1200, height = 500, title = "Stock Analysis Daily Box-Plot", sizing_mode="scale_width")

h = 12*3600*1000
print(df.index, df.High, df.Low)
p.segment(df.index, df.High, df.index, df.Low, color="black")
p.rect(df1.index, (df1.Open + df1.Close)/2, h, abs(df1.Open - df1.Close), fill_color="#28df99", line_color="black")
p.rect(df2.index, (df2.Open + df2.Close)/2, h, abs(df2.Open - df2.Close), fill_color="#ffbb91", line_color="black")

script1, div1 = components(p)
cdn_js = CDN.js_files
cdn_css = CDN.css_files

#output_file("Stock.html")
#show(p)
```
